# Neo4j CQL - Introduction

<img src="images/Neo4j.png" width="30%">

CQL signifie Cypher Query Language. Comme un RDBMS a le langage de requête SQL, Neo4j a CQL comme langage de requête.

#### Neo4j CQL
<ul>
    <li>Est un langage de requête pour Neo4j Graph Database.</li>
    <li>Est un langage déclaratif de correspondance de modèles.</li>
    <li>Possède une syntaxe proche du langage SQL.</li>
    <li>La syntaxe est très simple et dans un format lisible par l'homme.</li>
</ul>

#### Comme  SQL
<ul>
    <li>Neo4j CQL a des commandes pour effectuer des opérations de base de données.</li>
<li>Neo4j CQL prend en charge de nombreuses clauses telles que WHERE, ORDER BY, etc., pour écrire facilement des requêtes très complexes.</li>
<li>Neo4j CQL prend en charge certaines fonctions telles que celles de manipulation de String, Aggregation. En plus, il prend également en charge certaines fonctions de relation.</li>
</ul>

#### Clauses Neo4j CQL
Voici les clauses de lecture de Neo4j Cypher Query Language :
<table>
    <tr><td>Read Clauses</td><td>Usage</td></tr>
    <tr><td>MATCH</td><td>Cette clause est utilisée pour rechercher les données avec un modèle spécifié.</td></tr>
    <tr><td>OPTIONAL MATCH </td><td>C'est la même chose que match, la seule différence étant qu'il peut utiliser des valeurs nulles en cas de parties manquantes du motif.</td></tr>
    <tr><td>WHERE </td><td>Cette clause est utilisé pour ajouter des conditions aux requêtes CQL.</td></tr>
    <tr><td>START </td><td>Cette clause est utilisée pour trouver les points de départ au moment de l'affichage.</td></tr>
     <tr><td>LOAD CSV</td><td>Cette clause est utilisée pour importer des données à partir de fichiers CSV.</td></tr>
</table>

<p>Voici les clauses d'écriture de Neo4j Cypher Query Language :</p>

<table>
    <tr><td>Write Clause</td><td>Usage</td></tr>
    <tr><td>CREATE</td><td>Cette clause est utilisée pour créer des nœuds, des relations et des propriétés.</td></tr>
    <tr><td>MERGE</td><td>Cette clause vérifie si le modèle spécifié existe dans le graphe. Sinon, il crée le motif.</td></tr>
    <tr><td>SET</td><td>Cette clause est utilisée pour mettre à jour les étiquettes sur les nœuds, les propriétés sur les nœuds et les relations.</td></tr>
    <tr><td>DELETE</td><td>Cette clause est utilisée pour supprimer des nœuds et des relations ou des chemins, etc. du graphe.</td></tr>
    <tr><td>REMOVE</td><td>Cette clause est utilisée pour supprimer des propriétés et des éléments des nœuds et des relations.</td></tr>
    <tr><td>FOREACH</td><td>Cette classe est utilisée pour mettre à jour les données dans une liste.</td></tr>
    <tr><td>CREATE UNIQUE</td><td>En utilisant les clauses CREATE et MATCH, vous pouvez obtenir un modèle unique en faisant correspondre le modèle existant et en créant celui qui manque.</td></tr>
    <tr><td>Importing CSV files with Cypher</td><td>À l'aide de Load CSV, vous pouvez importer des données à partir de fichiers .csv.</td></tr>
    
</table>

<p>Voici les clauses générales de Neo4j Cypher Query Language :</p>

<table>
    <tr><td>General Clauses</td><td>Usage</td></tr>
    <tr><td>RETURN</td><td>Cette clause est utilisée pour définir ce qu'il faut inclure dans le jeu de résultats de la requête.</td></tr>
    <tr><td>ORDER BY</td><td>Cette clause est utilisée pour organiser la sortie d'une requête dans l'ordre. Il est utilisé avec les clauses RETURN ou WITH.</td></tr>
    <tr><td>LIMIT</td><td>Cette clause est utilisée pour limiter les lignes du résultat à une valeur spécifique.</td></tr>
    <tr><td>SKIP</td><td>Cette clause est utilisée pour définir à partir de quelle ligne commencer en incluant les lignes dans la sortie.</td></tr>
    <tr><td>WITH</td><td>Cette clause est utilisée pour enchaîner les parties de la requête.</td></tr>
    <tr><td>UNWIND</td><td>Cette clause est utilisée pour développer une liste en une séquence de lignes.</td></tr>
    <tr><td>UNION</td><td>Cette clause est utilisée pour combiner le résultat de plusieurs requêtes.</td></tr>
    <tr><td>CALL</td><td>Cette clause permet d'invoquer une procédure déployée dans la base de données.</td></tr>
</table>

#### Types de données Neo4j CQL

    <tr><td>short</td><td>Il est utilisé pour représenter des entiers 16 bits.</td></tr>
    <tr><td>int</td><td>Il est utilisé pour représenter des entiers 32 bits.</td></tr>
    <tr><td>long</td><td>Il est utilisé pour représenter des entiers 64 bits.</td></tr>
    <tr><td>float</td><td>t est utilisé pour représenter des nombres à virgule flottante 32 bits.</td></tr>
    <tr><td>double</td><td>t est utilisé pour représenter des nombres à virgule flottante 64 bits.</td></tr>
    <tr><td>char</td><td>Il est utilisé pour représenter des caractères 16 bits.</td></tr>
    <tr><td>String</td><td>Il est utilisé pour représenter des chaînes.</td></tr>
</table>
